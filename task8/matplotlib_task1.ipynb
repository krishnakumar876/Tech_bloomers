Matplotlib  basics


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)

plt.show
```

```python
plt.xlabel("X alues")
plt.ylabel('Y values')
plt.title('Graph of $y = x^2$')

plt.plot(x, y)
plt.show()
```

```python
x1 = np.linspace(-np.pi, np.pi, 25)
x2 = np.linspace(-np.pi, np.pi, 25)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
x1
```

```python
y1
```

```python
x2
```

```python
y2
```

```python
plt.plot(x1,y1,x2,y2)
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(x1,y1, 'r-.')
plt.plot(x2,y2, 'b--')
plt.show()
```
